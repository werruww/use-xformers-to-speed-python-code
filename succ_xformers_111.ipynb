```python
# Copyright (c) Facebook, Inc. and its affiliates. All rights reserved.
#
# This source code is licensed under the BSD license found in the
# LICENSE file in the root directory of this source tree.
"""
This file contains biases that can be used as the `attn_bias` argument in
:attr:`xformers.ops.memory_efficient_attention`.
Essentially, a bias is a Tensor which will be added to the ``Q @ K.t`` before
computing the ``softmax``.


The goal of having custom made classes (instead of dense tensors) is that
we want to avoid having to load the biases from memory in the kernel, for
performance reasons. We also want to be able to know before-hand which
parts of the attention matrix we will need to compute (eg causal masks).


Some very common biases are LowerTriangularMask and BlockDiagonalMask.
"""

import math
from dataclasses import dataclass
from typing import (
    Any,
    ClassVar,
    Iterable,
    List,
    Optional,
    Sequence,
    Tuple,
    Type,
    Union,
    cast,
)

import torch


class AttentionBias:
    """Base class for a custom bias that can be applied \
        as the attn_bias argument in
    :attr:`xformers.ops.memory_efficient_attention`.

    That function has the ability to add a tensor, the
    attention bias, to the QK^T matrix before it is used
    in the softmax part of the attention calculation.
    The attention bias tensor with shape
    (B or 1, n_queries, number of keys)
    can be given as the attn_bias input.
    The most common use case is for an attention bias is
    to contain only zeros and negative infinities, which forms
    a mask so that some queries only attend to some keys.

    Children of this class define alternative things which can
    be used as the attn_bias input to define an attention bias which
    forms such a mask, for some common cases.

    When using an :attr:`xformers.ops.AttentionBias`
    instead of a :attr:`torch.Tensor`, the mask matrix does
    not need to be materialized, and can be
    hardcoded into some kernels for better performance.

    See:

    - :attr:`xformers.ops.fmha.attn_bias.LowerTriangularMask`
    - :attr:`xformers.ops.fmha.attn_bias.LowerTriangularFromBottomRightMask`
    - :attr:`xformers.ops.fmha.attn_bias.LowerTriangularMaskWithTensorBias`
    - :attr:`xformers.ops.fmha.attn_bias.BlockDiagonalMask`
    - :attr:`xformers.ops.fmha.attn_bias.BlockDiagonalCausalMask`

    """

    HOLDS_DENSE_TENSOR = False

   def materialize(
        self,
        shape: Tuple[int, ...],
        dtype: torch.dtype = torch.float32,
        device: Union[str, torch.device] = "cpu",
    ) -> torch.Tensor:
        """
        Materializes the bias as a `torch.Tensor`. This is very slow
        and we don't attempt to make it fast. Only use for debugging/testing.

        Shape should be like `[*, q_seqlen, k_seqlen]`
        """
        raise NotImplementedError()



def _get_default_bias_device(device: Optional[torch.device] = None) -> torch.device:
    if device is None:
        if torch.cuda.is_available():
            return torch.device("cuda")
        return torch.device("cpu")
    return device
```

```python
import torch
import time
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import xformers.ops as xops

# تحميل النموذج و Tokenizer
model_name = "openai-community/gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# نقل النموذج إلى GPU إذا كان متاحًا
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# نص الإدخال
input_text = "ما هو الغرض من الذكاء الاصطناعي؟"
inputs = tokenizer(input_text, return_tensors="pt").to(device)

# وظيفة لتعديل طبقة الانتباه لاستخدام XFormers
def replace_attention_with_xformers(model):
    for layer in model.transformer.h:
        # حفظ الانتباه الأصلي
        original_attention = layer.attn

        # تعريف وظيفة الانتباه المخصصة باستخدام XFormers
        def xformers_attention(hidden_states, attention_mask=None, head_mask=None, **kwargs):  # Add **kwargs here
            # تحضير query, key, value
            query = original_attention.q_proj(hidden_states)
            key = original_attention.k_proj(hidden_states)
            value = original_attention.v_proj(hidden_states)

            # Change shape to match XFormers
            batch_size, seq_len, _ = hidden_states.shape
            num_heads = original_attention.num_heads
            head_dim = original_attention.head_dim

            query = query.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
            key = key.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
            value = value.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)

            # Apply XFormers
            attn_output = xops.memory_efficient_attention(query, key, value)

            # Restore original shape
            attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, -1)

            # Apply final layer
            attn_output = original_attention.out_proj(attn_output)
            return attn_output

        # استبدال الانتباه الأصلي بالانتباه المخصص
        layer.attn.forward = xformers_attention

# الاستدلال بدون XFormers
print("الاستدلال بدون XFormers:")
start_time = time.time()
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=50)
elapsed_time = time.time() - start_time
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
print(f"الوقت المستغرق: {elapsed_time:.2f} ثانية\n")

# تعديل النموذج لاستخدام XFormers
replace_attention_with_xformers(model)

# الاستدلال مع XFormers
print("الاستدلال مع XFormers:")
start_time = time.time()
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=50)
elapsed_time = time.time() - start_time
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
print(f"الوقت المستغرق: {elapsed_time:.2f} ثانية")
```

```python
import time
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import xformers.ops as xops

# تحميل النموذج وtokenizer بدون استخدام XFormers
model_name = "openai-community/gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# جملة نачائية للاستدلال عليها
input_text = "ما هو الغرض من الذكاء الاصطناعي؟"

# ترميز الجملة وتحويلها إلى مدخلات للنموذج
inputs = tokenizer.encode(input_text, return_tensors="pt")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
inputs = inputs.to(device)

# بدء العداد الزمني قبل الاستدلال
start_time = time.time()

# استدلال النموذج بدون تعديلات XFormers
outputs = model.generate(inputs, max_length=50)

# حساب الوقت المستغرق
end_time = time.time()
elapsed_time = end_time - start_time

# فك ترميز الإخراج وطباعته
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("الإجابة بدون تعديلات XFormers:")
print(generated_text)
print(f"الوقت المستغرق: {elapsed_time:.2f} ثانية")

# الآن، نعدل النموذج لاستخدام memory_efficient_attention
class GPT2AttentionWithXFormers(torch.nn.Module):
    def __init__(self, original_attention):
        super().__init__()
        self.original_attention = original_attention

    def forward(
        self,
        hidden_states: torch.Tensor,
        attention_mask: Optional[torch.FloatTensor] = None,
        head_mask: Optional[torch.FloatTensor] = None,
        encoder_hidden_states: Optional[torch.FloatTensor] = None,
        encoder_attention_mask: Optional[torch.FloatTensor] = None,
        past_key_value: Optional[Tuple[Tensor, Tensor]] = None,
        output_attentions: bool = False,
    ) -> Tuple[torch.Tensor]:
        query = self.original_attention.query(hidden_states)
        key = self.original_attention.key(hidden_states)
        value = self.original_attention.value(hidden_states)

        # استخدام memory_efficient_attention من XFormers
        attention_output = xops.memory_efficient_attention(
            query, key, value,
            attn_bias=None if attention_mask is None else -1e4 * (1.0 - attention_mask),
            p=self.original_attention.dropout.p if self.training else 0.0,
            scale=1 / torch.sqrt(torch.tensor(self.original_attention.embed_dim).to(query)),
        )

        # إعادة تشكيل الإخراج
        attention_output = attention_output.permute(0, 2, 1, 3)
        attention_output = attention_output.reshape(
            attention_output.size(0), attention_output.size(1), -1
        )
        output = self.original_attention.dense(attention_output)

        if output_attentions:
            return (output, None)  # لا نملك اطلاعات الإنتباه من XFormers

        return (output,)


# تعديل طبقة المحولات في النموذج
for layer in model.transformer.h:
    layer.attn = GPT2AttentionWithXFormers(layer.attn)

# بدء العداد الزمني قبل الاستدلال مع XFormers
start_time = time.time()

# استدلال النموذج باستخدام memory_efficient_attention
outputs_xformers = model.generate(inputs, max_length=50)

# حساب الوقت المستغرق
end_time = time.time()
elapsed_time_xformers = end_time - start_time

# فك ترميز الإخراج وطباعته
generated_text_xformers = tokenizer.decode(outputs_xformers[0], skip_special_tokens=True)
print("الإجابة باستخدام memory_efficient_attention من XFormers:")
print(generated_text_xformers)
print(f"الوقت المستغرق: {elapsed_time_xformers:.2f} ثانية")
```

```python
import time
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
import xformers.ops as xops


# تحميل النموذج و Tokenizer
model_name = "openai-community/gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)


# نقل النموذج إلى GPU إذا كان متاحًا
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)


# نص الإدخال
input_text = "ما هو الغرض من الذكاء الاصطناعي؟"
inputs = tokenizer(input_text, return_tensors="pt").to(device)


# وظيفة لتعديل طبقة الانتباه لاستخدام XFormers
def replace_attention_with_xformers(model):
    for layer in model.transformer.h:  # h for GPT2 models
        # Check if the layer has an SDPA attention mechanism
        if hasattr(layer.attn, "attn"):
            original_attention = layer.attn.attn
        else:
            # Continue with the previous logic if it's not an SDPA layer
            original_attention = layer.attn

        # Check if the original attention mechanism has `q_proj`, etc.
        if hasattr(original_attention, "q_proj"):
            # Define a custom attention function using XFormers
            def xformers_attention(hidden_states, attention_mask=None, head_mask=None, **kwargs):
                # Access `q_proj`, `k_proj`, and `v_proj` attributes
                query = original_attention.q_proj(hidden_states)
                key = original_attention.k_proj(hidden_states)
                value = original_attention.v_proj(hidden_states)

                # Change the shape to match XFormers
                batch_size, seq_len, _ = hidden_states.shape
                num_heads = original_attention.num_heads
                head_dim = original_attention.head_dim

                query = query.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
                key = key.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
                value = value.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)

                # Apply XFormers
                attn_output = xops.memory_efficient_attention(query, key, value)

                # Restore the original shape
                attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, -1)

                # Apply the final layer
                attn_output = original_attention.out_proj(attn_output)

                return attn_output

            # Replace the original attention mechanism with the custom one
            original_attention.forward = xformers_attention
        else:
            # Skip this layer if it does not have the necessary attributes for XFormers
            print(f"Skipping layer {layer} as it does not have q_proj, k_proj, and v_proj.")

# الاستدلال بدون XFormers
print("الاستدلال بدون XFormers:")
start_time = time.time()
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=50)
elapsed_time = time.time() - start_time
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
print(f"الوقت المستغرق: {elapsed_time:.2f} ثانية\n")


# تعديل النموذج لاستخدام XFormers
replace_attention_with_xformers(model)


# الاستدلال مع XFormers
print("الاستدلال مع XFormers:")
start_time = time.time()
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=50)
elapsed_time = time.time() - start_time
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
print(f"الوقت المستغرق: {elapsed_time:.2f} ثانية")
```

Scaled dot product attention
torch.nn.functional.scaled_dot_product_attention (SDPA) is an optimized and memory-efficient attention (similar to xFormers) that automatically enables several other optimizations depending on the model inputs and GPU type. SDPA is enabled by default if you’re using PyTorch 2.0 and the latest version of 🤗 Diffusers, so you don’t need to add anything to your code.

However, if you want to explicitly enable it, you can set a DiffusionPipeline to use AttnProcessor2_0:

‏torch.nn.functional.scaled_dot_product_attention‏‏ (SDPA) هو اهتمام محسن وفعال للذاكرة (على غرار xFormers) يتيح تلقائيا العديد من التحسينات الأخرى اعتمادا على مدخلات النموذج ونوع وحدة معالجة الرسومات. يتم تمكين SDPA افتراضيا إذا كنت تستخدم PyTorch 2.0 وأحدث إصدار من 🤗 Diffusers ، لذلك لا تحتاج إلى إضافة أي شيء إلى التعليمات البرمجية الخاصة بك.‏
